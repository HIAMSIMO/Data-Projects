# **Scraping Data from a Real Website Project**
This script scrapes data from a Wikipedia page and saves it as a CSV file.

# Importing required libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Scrap the data from target URL

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

Finding all the headers in the table and Creating a list of headers

```python
table = soup.find_all('table')[1]
world_titles = table.find_all('th')
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

# Creating a DataFrame from an HTML table by setting column names to headers and iterating through each row for data.

```python
df = pd.DataFrame(columns = world_table_titles)

column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

df
```

# **Save the DataFrame to a CSV file**

```python
df.to_csv(r'Data.csv',index= False)
```
